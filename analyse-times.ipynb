```python
import wkr
import numpy as np
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = [12,8]
```

```python
def loadtimes(hidraw_file):
    lines = wkr.lines(hidraw_file)
    times = [float(line.split()[1]) for line in lines if line.startswith('E: ')]
    return np.array(times)

tp_hidraw = np.diff(loadtimes('touchpad.hid'))
um_hidraw = np.diff(loadtimes('usbmouse.hid'))
```

```python
plt.hist(tp_hidraw[tp_hidraw < 0.1], range=(0.,0.1))
plt.xlabel('Time difference between events (seconds)')
plt.ylabel('Number of events')
plt.title('Recording touchpad events with hid-recorder')
```

```python
plt.hist(um_hidraw[um_hidraw < 0.1], range=(0.,0.1))
plt.xlabel('Time difference between events (seconds)')
plt.ylabel('Number of events')
plt.title('Recording USB mouse events with hid-recorder')
```

```python
tp_evemu = np.diff(loadtimes('touchpad-recording.evemu.gz'))
um_evemu = np.diff(loadtimes('usbmouse-recording.evemu.gz'))
```

```python
plt.hist(tp_evemu[tp_evemu < 0.2], range=(0.,0.2))
plt.xlabel('Time difference between events (seconds)')
plt.ylabel('Number of events')
plt.title('Recording touchpad events with evemu')
```
